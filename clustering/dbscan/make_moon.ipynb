# Clustering using DBSCAN

### Import libraries

```python
import random 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import warnings 
warnings.filterwarnings(action='ignore')
```

```python
from sklearn.datasets import make_moons
X,y = make_moons(n_samples=100, random_state=1)
```

```python
df = pd.DataFrame(X)
```

```python
df.head()
```

```python
plt.scatter(X[:,0],X[:,1])
```

## Using k-means

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
```

```python
centers = kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1], c=labels.astype(float))
plt.xlabel('Feature1', fontsize=14)
plt.ylabel('Feature2', fontsize=14)
# draw centers 
plt.scatter(centers[:,0], centers[:,1], c="red", marker="s")
plt.show()
```

## Using AgglomerativeClustering 

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model  =  AgglomerativeClustering(n_clusters=2)
model.fit(df)
```

```python
labels = model.labels_
```

```python
plt.scatter(X[:,0],X[:,1], c=labels.astype(float))
plt.xlabel('Feature1', fontsize=14)
plt.ylabel('Feature2', fontsize=14)
plt.show()
```

## Using DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.25, min_samples=5)
dbscan.fit(X)
```

```python
df['labels'] = dbscan.labels_
```

```python
df.labels.value_counts()
```

```python
plt.scatter(X[:,0], X[:,1], c=df.labels.astype(float))
plt.show()
```
