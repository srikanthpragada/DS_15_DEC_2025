# Clustering using DBSCAN

### Import libraries

```python
import random 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

### Load Data From CSV File  

```python
df = pd.read_csv("customers.csv")
df.head()
```

```python
df.shape
```

### Pre-processing

```python
df = df [['Age','Income']]
df.head()
```

```python
#Normalize data
from sklearn.preprocessing import scale
df['Age'] = scale(df['Age'])
df['Income'] = scale(df['Income'])
df.head()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=.5, min_samples=3)
dbscan.fit(df)
```

```python
labels = dbscan.labels_
labels[:100]
```

```python
pd.Series(labels).value_counts()
```

```python
plt.scatter(df['Age'], df['Income'], c=labels.astype(float))
plt.xlabel('Age', fontsize=14)
plt.ylabel('Income', fontsize=14)
plt.show()
```

```python
len(dbscan.components_)
```

```python
len(dbscan.core_sample_indices_)
```

```python
core_points = df.loc[dbscan.core_sample_indices_,:]
core_labels = labels[dbscan.core_sample_indices_]
```

```python
noise_index =[]
for idx in df.index:
    if labels[idx] == -1:
           noise_index.append(idx)
```

```python
len(noise_index)
```

```python
noise_points = df.loc[noise_index,:]
```

```python
len(noise_points)
```

```python
# Plot core points
plt.scatter(core_points['Age'], core_points['Income'], c=core_labels.astype(float))
plt.scatter(noise_points['Age'], noise_points['Income'], c='red', marker='*')
plt.show()
```
